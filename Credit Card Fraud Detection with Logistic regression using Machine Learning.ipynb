```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

This dataset is higly unbalanced

0---> Normal Transacion
1---> Fraudulent Transaction

```python
# Sperating the data for analysis

legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class ==1]
```

```python
# shape

print(legit.shape)
print(fraud.shape)
```

```python
# statistical Measures

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

Dealing wih the unbalanced data

Undersampling- Build a sample dataset containing similar distribution of normal and fraudulent transactions

No. of fraudulent transactions ---> 492

```python
legit_sample = legit.sample(n=492)
```

concatenating 2 dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
new_dataset.head()
```

```python
# new Distribution

new_dataset['Class'].value_counts()
```

```python
# Checking and comparing the new mean to old mean

new_dataset.groupby('Class').mean()
```

Splitting the features and targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X
```

```python
Y
```

Splitting the data into training and test Data

```python
X_train, X_test, Y_train,Y_test = train_test_split(X, Y, test_size=0.2, random_state=2, stratify=Y)
```

```python
X.shape, X_train.shape, X_test.shape
```

Model Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Evaluation and Accuracy

```python
# Training Data

training_data = model.predict(X_train)
training_data_accuracy = accuracy_score(training_data,Y_train)

print('The accuracy score:', training_data_accuracy)
```

```python
# Test Data

test_data = model.predict(X_test)
test_data_accuracy = accuracy_score(test_data,Y_test)

print('The accuracy score:', test_data_accuracy)
```
